```python
sample_interval = 0.2
samples = 100 # you don't necessarily have to use a power of 2
time = np.linspace(-10, samples*sample_interval, samples)
signal_frequency = 0.5 # 0.5 Hz

def signal(time):
    signal=[]
    for i in range(len(time)):

        if time[i]>=0:

           y= np.cos(time[i]* np.pi*signal_frequency)
        else:
           y=0
        signal.append(y)
    return signal
signal = signal(time)
print(signal)
```

```python
signal_spectrum = np.fft.fftshift(np.fft.fft((signal)))
freqs = np.fft.fftshift(np.fft.fftfreq(samples, d=sample_interval))
plt.figure(figsize=(10,5))
plt.plot(freqs, np.abs(signal_spectrum),'g')  # in kHz
plt.xlim(-5, 5)
plt.ylim(-2.0, 40)
```

## Twice Low-pass filter

```python
#https://stackoverflow.com/questions/25191620/creating-lowpass-filter-in-scipy-understanding-methods-and-units
import numpy as np
from scipy.signal import butter, lfilter, freqz
import matplotlib.pyplot as plt
import plotly.graph_objects as go

def butter_lowpass(cutoff, fs, order):
    return butter(order, cutoff, fs=fs, btype='low', analog=False)

def butter_lowpass_filter(data, cutoff, fs, order):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    y = lfilter(b, a, y[::-1])[::-1]
    return y

# Filter requirements.
order = 7
fs = 30.0       # sample rate, Hz
cutoff = 3.667  # desired cutoff frequency of the filter, Hz

# Get the filter coefficients
b, a = butter_lowpass(cutoff, fs, order)

# Plot the frequency response.
w, h = freqz(b, a, fs=fs, worN=8000)
fig = go.Figure()
fig.add_trace(go.Scatter(x=w, y=np.abs(h), line =  dict(shape =  'spline',color='green'),name='Butterworth Lowpass Filter Frequency Response'))
fig.add_trace(go.Scatter(x=[cutoff], y=[0.5*np.sqrt(2)],line_shape='linear',name='Cut-Off Frequency'))
fig.add_vline(x=3.667, line_width=3, line_dash="dash", line_color="red")
fig.show()

# Demonstrate the use of the filter.
# First make some data to be filtered.
T = 5.0         # seconds
n = int(T * fs) # total number of samples
t = np.linspace(0, T, n, endpoint=False)
data = np.cos(1.2*2*np.pi*t) + np.cos(9*2*np.pi*t)

# Filter the data, and plot both the original and filtered signals.
y = butter_lowpass_filter(data, cutoff, fs, order)

# Filter the data, and plot both the original and filtered signals.
fig = go.Figure()
fig.add_trace(go.Scatter(
            y = data,
            line =  dict(shape =  'spline',color='green' ),
            name = 'signal with noise'
            ))
fig.add_trace(go.Scatter(
            y = y,
            line =  dict(shape =  'spline',color='red' ),
            name = 'Twiced-filtered signal',

            ))
fig.show()
```

```python
#https://github.com/mubeta06/python/blob/master/signal_processing/sp/multirate.py

def downsample(s, n, phase=0):
    """Decrease sampling rate by integer factor n with included offset phase.
    """
    return s[phase::n]

def decimate(s, r, n=None, fir=False):

    if fir:
        if n is None:
            n = 30
        b = signal.firwin(n, 1.0/r)
        a = 1
        f = signal.lfilter(b, a, s)
    else: #iir
        if n is None:
            n = 8
        b, a = signal.cheby1(n, 0.05, 0.8/r)
        f = signal.filtfilt(b, a, s)
    return downsample(f, r)
```
