# 1)Convolution
  Assume x(n) = u(n)-u(n-10) and h(n)= an u(n), with a=0.5.
a) Find the theoretical closed-form expression for y(n) = x(n)*h(n)
b) Plot h(n)
c) Plot x(n) and y(n).

## u[n]

```python
import numpy as np
import matplotlib.pyplot as plt

n = range(-2, 6, 1)
y = []
for i in range(len(n)):
    temp = (1 if n[i]>=0 else 0)
    y.append(temp)
print(n)
print(y)
#plotting the graph
plt.stem(n, y)
plt.axis([-2.1, 5.1, -0.1, 1.2])
plt.show()
```

## h[n]

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.5
n = range(0, 20, 1)
#n_u = range(0, 26, 1)
h = []

for i in range(len(n)):
    temp_u = (1 if n[i]>=0 else 0)
    temp = (a**n[i])* temp_u
    h.append(temp)

#Plotting the graph
plt.stem(h)
plt.axis([-0.5, 19.1, -0.1, 1.1])
plt.xlabel('  n  ')
plt.ylabel('Amplitude  ')
plt.title('h(n) for a=0.5')
plt.grid(True)
plt.show()
```

## x[n] = u[n] - u[n-N]

```python
import numpy as np
import matplotlib.pyplot as plt

n = range(0, 20, 1)
x = []
for i in range(len(n)):
    temp = (1 if n[i]>=0 else 0)
    temp2 = (1 if n[i]-10>=0 else 0)
    x.append((temp-temp2))

print(n)
print(x)
#plotting the graph
plt.stem(n, x)
plt.axis([-0.5, 19.1, -0.1, 1.2])
plt.ylabel('Amplitude  ')
plt.title('x(n) = u[n] - u[n-10]')
plt.show()
```

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

y = signal.convolve(x,h)
print(y)
fig, ax1 = plt.subplots(1,1,figsize=(30,15))
n = range(0, 39, 1)
ax1.stem(n, y, 'g')
plt.title('y[n]')
```

# 3) Approximation of Ideal Lowpass Filter

 The ideal lowpass filter in Example 2.18 is not implementable. A simple approximation can be achieved by truncating the impulse response. Assume a noncausal L-tap (L is odd) FIR filter is obtained by such a truncation. To make this filter causal, one can shift the impulse response to the right by (L-1)/2 samples, giving

a) Assuming ωC=0.2π, compute and plot the amplitude and phase responses of this causal FIR filter for L=19 and L=101.
b) What are the advantages and disadvantages of increasing L?
c) Plot the amplitude and phase responses of (-1)nhFIR(n) and compare with the results in a).

## a,b)

```python
#https://mpastell.com/pweave/_downloads/FIR_design_minted.pdf

from pylab import *
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
import math

#Plot frequency and phase response
def mfreqz(b,a=1):
    w,h = signal.freqz(b,a)
    h_dB = 20 * log10 (abs(h))
    fig, ax1 = plt.subplots(1,1,figsize=(10,5))
    h_Phase = unwrap(arctan2(imag(h),real(h)))
    plot(w/max(w),h_Phase,'g')
    ylabel("Phase (radians)")
    xlabel(r"Normalized Frequency (x$\pi$rad/sample)")
    title(r"Phase response")
    subplots_adjust(hspace=0.5)

#Plot step and impulse response
def impz(b,n,a):
   l = len(b)
   impulse = repeat(0.,l); impulse[0] =1.
   x = arange(0,l)
   response = signal.lfilter(b,a,impulse)
   fig, ax1 = plt.subplots(1,1,figsize=(10,5))
   step = cumsum(response)
   ax1.stem(x, step,'g')
   ylabel("Amplitude")
   xlabel(r"n (samples)={:.0f},".format(n))
   subplots_adjust(hspace=0.5)
```

```python
from pylab import *
import scipy.signal as signal

n = 19
a = signal.firwin(n, cutoff = (math.pi)*0.2, window = "hamming")
#Frequency and phase response
mfreqz(a)
show()
#Impulse and step response
figure(2)
m=n-1//2
if n>m :
  h=(math.sin(0.2 *(n-m)))//(math.pi*(n-m))
else :
  h=1
impz(a,n,h)
show()
```
