### Phase Function


```python
def warpingphase(w, a1,a2):
  #phase of allpass zero/pole
  theta1=np.angle(a1)
  theta2=np.angle(a2)
  #magnitude of allpass zero/pole
  r1=np.abs(a1)
  r2=np.abs(a2)
  wy = (-w*2) -(2* np.arctan((r1*np.sin(w-theta1))/(1-r1*np.cos(w-theta1))))-(2* np.arctan(r2*np.sin(w+theta2))/(1-r2*np.cos(w+theta2)))
  return wy
```

```python
import matplotlib.pyplot as plt

w=np.arange(0,3.14,0.01)
plt.figure()
plt.plot(w,warpingphase(w,a1,a2))
plt.grid()
plt.xlabel("Normalized Frequency")
plt.ylabel('Phase Angel')
```

## Kaiser Windowing

### M=24

```python
from scipy import signal
import numpy as np
import numpy as np
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt
np.seterr(divide = 'ignore')
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.windows.kaiser.html

M=24
window = signal.windows.kaiser(24, beta=2.44)

#https://dsp.stackexchange.com/questions/50563/phase-response-of-a-fir-hilbert-transformer
N = M//2
b = [2*np.sin(np.pi*n/2)**2/(np.pi*n) for n in range(1, N)]
b = np.append(-1*np.flip(b, 0), np.append([0], b))
b= np.append(b,2.864330194332735e-32)
b=b*window

print("window",window.shape)
print("b",b.shape)
w, h = signal.freqz(b)
h = h*np.exp(1j*w*N)
plt.stem(b)
plt.ylabel('Amplitude  ')
plt.title('h(n)')
plt.show()
fig, ax1 = plt.subplots(1,1,figsize=(7,5))
plt.title('filter frequency response with Kaiser window M=24')
plt.plot(w, 20 * np.log10(abs(h)), 'g')
plt.ylabel('Amplitude [dB]', color='g')
plt.xlabel('Normalized Frequency [rad/sample]', color='black')
ax1 = ax1.twinx()
angles = np.angle(h)
plt.plot(w, angles, 'b')
plt.ylabel('Phase difference (radians)', color='b')
plt.xlabel('Normalized Frequency [rad/sample]', color='black')
plt.show()
```
